# HDBSCAN clustering
  -
HDBSCAN (Hierarchical Density-Based Spatial Clustering of Applications with Noise) is a clustering algorithm that's particularly useful for discovering clusters of varying densities and shapes in data. Unlike traditional clustering methods, HDBSCAN doesn't require specifying the number of clusters beforehand and can handle noise effectively. It builds a hierarchy of clusters and identifies the most stable ones based on density. This makes it robust in handling complex datasets where clusters may have different sizes or densities. HDBSCAN's ability to automatically determine the number of clusters and its resilience to noise make it a valuable tool in exploratory data analysis and pattern recognition tasks.

```python
clusterer = hdbscan.HDBSCAN(min_cluster_size=10, metric='euclidean', cluster_selection_method='eom', min_samples = 20,
                            cluster_selection_epsilon=0.3)
clusterer.fit(umap_embeddings)
print(len(Counter(list(clusterer.labels_)).most_common()))
Counter(list(clusterer.labels_)).most_common()
```

```python
def pct_rank_qcut(series, n):
    """

      The provided function, pct_rank_qcut, appears to perform percentile rank quantile cuts on a given series. Here's a breakdown of its functionality:

      1. It takes two parameters: series, which is a pandas Series object containing the data to be processed, and n, which represents the number of quantiles or percentiles to divide the data into.
      2. It initializes a pandas Series called edges containing n+1 evenly spaced quantile boundaries ranging from 0 to 1.
      3. It defines a lambda function f that takes a value x and finds the index of the first element in edges that is greater than or equal to x.
      4. It calculates the percentile ranks of the values in the input series using the rank method with pct=1, which returns the percentage rank of each element in the series.
      5. It applies the lambda function f to each percentile rank calculated in step 4, effectively assigning each percentile rank to its corresponding quantile.
      6. It returns a new pandas Series where each element represents the quantile to which the corresponding percentile rank belongs.

      In summary, this function takes a series of numerical data and divides it into n quantiles based on percentile ranks, assigning each value in the series to its respective quantile.
    """
    edges = pd.Series([float(i) / n for i in range(n + 1)])
    f = lambda x: (edges >= x).argmax()
    return series.rank(pct=1).apply(f)

def create_index(a):
    """
      This Python function creates an index using the nmslib library for a given set of data points represented by the array a. Here's what each step does:

      1. nmslib.init(space='cosinesimil'): Initializes an index using the NMSLIB library, specifying the distance metric to be used as cosine similarity.
        Cosine similarity is often used when dealing with high-dimensional data, such as text or document embeddings, to measure the similarity between vectors regardless of their magnitude.
      2. index.addDataPointBatch(a): Adds the data points represented by the array a to the index in batches. This step prepares the data for efficient querying by the index.
      3. index.createIndex(): Creates the index structure based on the data points added to the index. This step builds the underlying data structure that facilitates fast similarity searches.
      4. return index: Returns the created index, which can then be used to perform nearest neighbor searches or other similarity-based queries efficiently.

      Overall, this function initializes an index, adds data points to it, creates the index structure, and returns the resulting index object for further use in similarity search tasks.
    """
    index = nmslib.init(space='cosinesimil')
    index.addDataPointBatch(a)
    index.createIndex()
    return index

def get_knns(index, vecs, k=3):
    """
    This function, named get_knns, appears to retrieve the k-nearest neighbors (KNNs) for a set of vectors using an index. Here's a breakdown:
    - index: It's likely an index structure (such as a spatial index or a similarity index) built from a collection of vectors. This index allows for efficient nearest neighbor queries.
    - vecs: This parameter represents a collection of vectors for which we want to find the k-nearest neighbors.
    - k: It specifies the number of nearest neighbors to retrieve. By default, it's set to 3.
    - num_threads: This parameter seems to specify the number of threads to be used for the nearest neighbor query. In this case, it's set to 8.
    The function utilizes the knnQueryBatch method of the index object to efficiently find the k-nearest neighbors for multiple vectors in vecs. It returns the results in a format that separates the indices of the nearest neighbors from their corresponding distances. The zip(*...) operation effectively transposes the result from a list of pairs (index, distance) to separate lists of indices and distances.
    """
    return zip(*index.knnQueryBatch(vecs, k=k,num_threads=8))

def fn_distance_metrics(df, column):
  """
  This function, fn_distance_metrics, computes distances between pairs of vectors in a DataFrame column using a nearest neighbor approach. Here's a breakdown of its steps:
  1. Create Nearest Neighbor Index: It creates a nearest neighbor index (nn_wvs) based on the vectors in the specified column of the DataFrame df.
  2. Get Nearest Neighbors: It then retrieves the k-nearest neighbors for each vector using the created index.
  3. Prepare Dataframes: It converts the indices and corresponding distances into pandas DataFrames (idxs and dists, respectively) while dropping the first column which is the original index.
  4. Concatenate DataFrames: It concatenates the indices and distances DataFrames column-wise (catted), so that each row represents a pair of vectors along with their distance.
  5. Reset Index and Rename Columns: It resets the index of the concatenated DataFrame and renames the columns appropriately (v1 for the first vector, v2 for the second vector, and dist for the distance).
  6. Return Result: Finally, it returns the concatenated DataFrame containing the indices of vector pairs and their corresponding distances.

  In summary, this function facilitates the computation of distances between pairs of vectors in a DataFrame column, providing a structured output for further analysis or processing.
  """
  nn_wvs = create_index(list(df[column]))
  to_frame = lambda x: pd.DataFrame(np.array(x)[:,1:])
  idxs, dists = map(to_frame, get_knns(nn_wvs, list(df[column]), k=10))

  catted = pd.concat([idxs.stack().to_frame('idx'),
                      dists.stack().to_frame('dist')], axis=1).reset_index().drop('level_1', axis=1).rename(columns={'level_0': 'v1', 'idx': 'v2'})
  return catted

# Exclude the cluster with name -1 --> Global outliers
list_clusters=list(df.hdbscan_cluster.unique())
list_clusters_analyze = [x for x in list_clusters if x not in [-1]]

# initializes an empty pandas DataFrame df_report
df_report = pd.DataFrame(columns=['hdbscan_cluster', 'template_flag', 'sample_size', 'sample_template'])
count = 0
df.reset_index(inplace=True, drop=True)

for clust in list_clusters_analyze:
  """
  For each cluster (clust) in the list list_clusters_analyze, it:
    1. Selects the subset of the DataFrame df where the 'hdbscan_cluster' column matches the current cluster.
    2. Computes distances between vectors in this subset using a function called fn_distance_metrics.
    3. Divides the distances into deciles and calculates the purity of each decile based on the number of unique data points it contains.
    4. Determines if a template flag should be set based on the highest purity decile having at least 10% of the total data points.
    5. If a template flag is set, selects a random email body from the data points in the highest purity decile as a sample template.
    6. Appends information about the cluster, template flag, sample size, and sample template to the df_report DataFrame.
  """
  print('cluster name: ', clust)
  template_flag = 0
  sample_size = 0
  sample = ''
  case_numbers = list()
  df_temp = df[df['hdbscan_cluster'] == clust]
  print(df_temp.columns)
  dist = fn_distance_metrics(df_temp, 'sent_embeddings')
  dist['decile_rank'] = pct_rank_qcut(dist.dist, 10)
  dist.sort_values('decile_rank', ascending=True, inplace=True)
  # dist_desc = pd.DataFrame(columns=['decile_rank', 'no_of_datapoints'])
  # for i in list(dist_desc.decile_rank.unique()):
  #   dist_sample = dist_desc[dist_desc.decile_rank == i].copy()
  #   new_df = {'decile_rank': i, 'no_of_datapoints': len(dist_sample.v1.unique())}
  #   dist_desc = dist_desc.append([new_df], ignore_index=True)

  dist_desc = pd.DataFrame(columns=['decile_rank', 'no_of_datapoints'])
  for i in list(dist.decile_rank.unique()):
    dist_sample = dist[dist.decile_rank == i].copy()
    new_row = {'decile_rank': i, 'no_of_datapoints': len(dist_sample.v1.unique())}
    new_df = pd.DataFrame([new_row])
    if len(dist_desc) == 0:
      dist_desc = new_df
    else:
      dist_desc = pd.concat([dist_desc, new_df], ignore_index=True)

  dist_desc['purity'] = (dist_desc.no_of_datapoints)/(dist_desc.no_of_datapoints.sum())
  print(dist_desc)
  if dist_desc.purity.max() >= 0.1:
    template_flag = 1
    max_decile_rank = list(dist_desc[dist_desc.purity == dist_desc.purity.max()].decile_rank)[0]
    sample = df_temp.iloc[[random.sample(list(dist[dist.decile_rank ==
                                                     max_decile_rank].v1.unique()), 1)[0]]].email_body.to_string(index=False)
    sample_size = len(list(dist[dist.decile_rank == max_decile_rank].v1.unique()))
    case_numbers = list(df_temp.iloc[dist[dist.decile_rank == max_decile_rank].v1.unique()].case_numbers)
  else:
    template_flag = 0
  # print(sample)
  new_row = {'hdbscan_cluster': clust,
            'template_flag': template_flag,
            'sample_size': sample_size,
            'sample_template': sample,
            'case_numbers': case_numbers}
  # df_report = df_report.append(new_row, ignore_index=True)
  df_report = pd.concat([df_report, pd.DataFrame([new_row])], ignore_index=True)
  # df_report = df_report.append({'hdbscan_cluster': clust,
  #                               'template_flag':template_flag,
  #                               'sample_size': sample_size,
  #                               'sample_template':sample,
  #                               'case_numbers':case_numbers},
  #                                ignore_index=True)
  count+=1

  if count % 100 == 0:
    print('No of data splits completed: ', count)
  gc.collect()
```

```python
def UncommonWords(A, B):
    """
      This Python function takes two strings A and B as input and returns a list of words that are uncommon between the two strings. Here's a breakdown of how it works:

      1. It initializes an empty dictionary count to store word counts.
      2. It iterates through each word in string A, splits it into individual words using split(), and updates the count of each word in the count dictionary.
      3. It does the same for string B, updating the word counts in the count dictionary.
      4. Finally, it returns a list comprehension that iterates through the keys (words) of the count dictionary and selects only those words whose count is equal to 1. These are the words that are present in only one of the input strings.

      In summary, the function identifies and returns the words that appear only once among the combined words of the two input strings.

    """
    # count will contain all the word counts
    count = {}

    # insert words of string A to hash
    for word in A.split():
        count[word] = count.get(word, 0) + 1

    # insert words of string B to hash
    for word in B.split():
        count[word] = count.get(word, 0) + 1

    # return required list of words
    return [word for word in count if count[word] == 1]

def is_date(string, fuzzy=False):
    """
    Return whether the string can be interpreted as a date.

    :param string: str, string to check for date
    :param fuzzy: bool, ignore unknown tokens in string if True
    """
    try:
        parse(string, fuzzy=fuzzy)
        return True

    except ValueError:
        return False
```

```python
df_final_template_report = pd.DataFrame(columns=['cluster', 'mail_template', 'sample_size', 'mail_sample_1', 'mail_sample_2'])

for idx, ele in enumerate(list_clusters):
  """
    This Python function appears to process a DataFrame (df_final_analysis) containing email data for analysis. Here's a breakdown of what it does:

    1. Initialization: It initializes an empty DataFrame (df_final_template_report) with specific column names.
    2. Loop through Clusters: It iterates over each cluster in list_clusters.
    3. Data Filtering and Analysis:
      - It filters df_final_analysis based on the current cluster.
      - It calculates the word length for each email in the cluster and filters out emails with word lengths below the 75th percentile.
      - It sorts the filtered emails by word length and selects the two shortest ones (list_res_summary).
    4. Text Processing:
      - It identifies uncommon words between the two selected emails.
      - It checks for salutations in the first line of the first email.
      - It categorizes words into different lists based on their characteristics like alphanumeric, numeric, date, URL, or email.
    5. Template Finalization:
      - It replaces identified codes, IDs, dates, URLs, and emails with placeholders in the first email template.
      - It cleans up other uncommon words in the first email template.
      - It stores the final email template, cluster ID, sample size, and sample emails into the df_final_template_report DataFrame.
    6. Iteration and Output: It repeats these steps for each cluster and updates the df_final_template_report DataFrame accordingly.

    In summary, this function processes email data per cluster, identifies common patterns, and generates a report of mail templates, their sample size, and samples with placeholders for identified entities like codes, IDs, dates, URLs, and emails.
  """

  df_temp = df_final_analysis[df_final_analysis.hdbscan_cluster == ele].copy()
  df_temp.loc[:, 'word_len'] = df_temp.email_body.apply(lambda x: len(x.split()))
  df_temp2 = df_temp[df_temp.word_len >= np.percentile(df_temp.word_len, 75)].copy()
  print(list(df_temp2.email_body))
  df_temp2.sort_values(by="word_len", ascending=True, inplace=True)
  list_res_summary = list(df_temp2.email_body[:2])

  if len(list_res_summary) > 1:
    ukn_words = UncommonWords(list_res_summary[0], list_res_summary[1])
    final_mail_template = list_res_summary[0]

    doc = nlp(final_mail_template)
    sents_list = []
    str_header = ''

    for sent in doc.sents:
        sents_list.append(sent.text)

    line1 = re.split(r"[~\r\n|,]+", sents_list[0])[0]

    if re.search(r'\b(hi|good|morning|day|hi|hello|hallo|dear|evening|customer|afternoon|sir|madam|mam)\b', line1.lower()):
      str_header = line1

    list_dates = []
    list_codes = []
    list_others = []
    list_ids = []
    list_url = []
    list_email = []
    dict_final = {}

    for word in ukn_words:
        if word.isalnum() and not word.isalpha() and not word.isdigit():
          list_codes.append(word)
        elif word.isdigit():
          list_ids.append(word)
        elif is_date(word):
          list_dates.append(word)
        elif validators.url(word) or validators.domain(word):
          list_url.append(word)
        elif validators.email(word):
          list_email.append(word)
        else:
          list_others.append(word)

    if len(list_codes) > 0:
      final_mail_template = re.compile('|'.join(map(re.escape, list_codes))).sub("<___Codes___>", final_mail_template)

    if len(list_ids) > 0:
      final_mail_template = re.compile('|'.join(map(re.escape, list_ids))).sub("<___IDs___>", final_mail_template)

    if len(list_dates) > 0:
      final_mail_template = re.compile('|'.join(map(re.escape, list_dates))).sub("<__Dates__>", final_mail_template)

    if len(list_url) > 0:
      final_mail_template = re.compile('|'.join(map(re.escape, list_url))).sub("<____url____>", final_mail_template)

    if len(list_email) > 0:
      final_mail_template = re.compile('|'.join(map(re.escape, list_email))).sub("<____email____>", final_mail_template)

    list_others_final = []
    for i in range(0, len(list_others)):
      ls_chk=[]
      ls_compare = [x for x in list_others if x != list_others[i]]
      if list_others[i].lower().strip('.') not in [x.lower().strip('.') for x in ls_compare]:
        list_others_final.append(list_others[i])

    if len(list_others_final) > 0:
      final_mail_template = re.compile('|'.join(map(re.escape, list_others_final))).sub("<_________>", final_mail_template)

    df_final_template_report.loc[idx, 'cluster'] = ele
    df_final_template_report.loc[idx, 'mail_template'] = str(final_mail_template)
    df_final_template_report.loc[idx, 'sample_size'] = len(df_temp)
    df_final_template_report.loc[idx, 'mail_sample_1'] = str(list_res_summary[0])
    df_final_template_report.loc[idx, 'mail_sample_2'] = str(list_res_summary[1])
```
